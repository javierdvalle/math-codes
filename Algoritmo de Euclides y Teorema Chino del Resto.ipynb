```python
def MCD(a, b):
    if a == 0:
        return b
    if b == 0:
        return a
    r = a%b
    q = int(a/b)
    assert a == b*q+r
    return MCD(b, r)

MCD(4, 12)
```

Explicación: 

Siendo A = **B**Q + **R**. A está formado por tanto por dos componentes: BQ y R.

- La idea es que basta con encontrar los números que dividan a las dos partes en negrita: a B y a R. Si hay alguno también divide a A. 
- Pero: ¿son esos *todos* los divisores comunes entre A y B? ¿Es condición suficiente dividir a B y R? Si: todo divisor de A,B tiene que dividir necesariamente a B (obvio) y a R (menos obvio). Probar que: Si x divide a A,B => entonces divide a R: Despejamos R = A - BQ, y vemos que x divide a la parte derecha y por tanto a la izda

Idse: 

- si R = 0, entonces solo tenemos el componente BQ que es divisible por todo B: de hecho, todos los divisores de B dividen a BQ
- si R != 0, entonces nos fijamos en R que es el que da problemas. Tenemos que mirar si R es divisible por al menos alguno de los divisores de B. Por eso miramos MCD(B, R). Si hay alguno, automáticamente ese número divide tanto al primer componente (es un divisor de B) como al segundo y por tanto divide a A.
