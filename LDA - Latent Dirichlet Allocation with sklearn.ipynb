```python
import wikipedia
import pandas as pd
import re
```

## Download Wikipedia articles

```python
articles = []
titles = wikipedia.random(50)
for title in titles:
    print('downloading:', title)
    try:
        article = wikipedia.page(title=title, auto_suggest=True, redirect=True)
        articles.append({'title': title, 'text': article.content, 'summary': article.summary, 'categories': article.categories})
    except Exception as e:
        print(e)

articles = pd.DataFrame(articles)
articles.head()
```
