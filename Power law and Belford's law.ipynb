# Power law and Benford's law

https://en.wikipedia.org/wiki/Power_law

https://en.wikipedia.org/wiki/Benford%27s_law


```python
import random
import numpy as np
import pylab as plt
import math
```

```python
def power_law(x, a, k):
    return a*(x**(-k))

N = 1000
a = 55
k = 2
x = np.linspace(1, 10, N)
y = power_law(x, a, k)

plt.plot(x, y)
```

```python
# tomando logaritmos queda lineal, sean cuales sean a y k

plt.plot(np.log(x), np.log(y))
```

## adding some random noise

```python
N = 1000
a = 55
k = 2
x = np.linspace(1, 10, N)
y = a*(x**(-k))  # power law
noise = np.random.uniform(0, 1, N)

plt.plot(x, y+noise)
```

```python
plt.plot(np.log(x), np.log(y+noise))
```

## Scale invariance 



```python
c = 2

plt.plot(np.log(x), np.log(y))
plt.plot(np.log(x), np.log(power_law(x*c, a, k)))

print(c, k)
print(c ** -k)
```

## Belford's law



```python
import seaborn as sns

a = 55
k = 10
x = np.random.uniform(0, 1, 30000)

y = power_law(x, a, k)

y = y*100000 # mover la coma a la derecha

first_digits = [int(str(int(e))[0]) for e in y if e >= 1]

first_digits[:10]
```

```python
sns.histplot(first_digits)
```

```python
from collections import Counter

freqs = Counter(first_digits)
total = sum(freqs.values())

for digit, count in sorted(freqs.items()):
    print(digit, count, count/total)
```
