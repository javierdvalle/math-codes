## Acertijo: ¿Cuál es el siguiente número en esta lista?


¿Cuál es el siguiente número en esta lista?:
4, 6, 12, 18, 30, 42, 60, 72, 102, 108, 138, ..


## Números primos

```python
def is_prime(n):
    return all(n%j for j in range(2,n))*(n>1)


for x in range(20):
    print(x, ':', True if is_prime(x) else False)
```

## Primos gemelos

https://en.wikipedia.org/wiki/Twin_prime

Hay algunos números primos que vienen por parejas, separados por un único número. El 3 y el 5 por ejemplo. Ahora bien, ¿hay infinitas de estas parejas? Esta pregunta tiene nombre, es la [conjetura de los números primos gemelos](https://es.wikipedia.org/wiki/Conjetura_de_los_n%C3%BAmeros_primos_gemelos) y es un caso concreto de la [conjetura de Polignac](https://en.wikipedia.org/wiki/Polignac%27s_conjecture)





```python
twin_primes = []

last_prime = 1
for x in range(10000):
    if is_prime(x):
        if x - last_prime == 2:
            twin_primes.append((last_prime, x))
        last_prime = x

twin_primes[:10] #first 10
```

## Conjetura de Goldbach

https://es.wikipedia.org/wiki/Conjetura_de_Goldbach


> Todo número par mayor que 2 puede escribirse como suma de dos números primos. Christian Goldbach (1742)


Ejemplos:
$
4 = 1+3 \\
8 = 3+5 \\
24 = 13+11 \\
52 = 23+29$

## Solución acertijo: Average of twin prime pairs

¿Cuál es el siguiente número en esta lista?:
4, 6, 12, 18, 30, 42, 60, 72, 102, 108, 138, ..

Más info aquí: http://oeis.org/A014574


```python
av_twins = [a+1 for (a, b) in twin_primes]

av_twins[:10] #first 10
```

## ¿Un caso especial de la Conjetura de Goldbach?

Todos los números en la lista anterior 4, 6, 12, 18, 30, 42, 60, 72, 102, 108, 138, ... parecen poder escribirse como suma de dos números anteriores en la lista. 

Ejemplos: 
$ 
18 = 12 + 6 \\
30 = 12 + 18 \\
42 = 30 + 12 \\
...
$

Podemos ver que esto ocurre, al menos, para todos los números primos gemelos menores que 100.000:

```python
for i in range(len(av_twins)):
    N = av_twins[i]
    print(i, N, N in [a + b for a in av_twins[:i] for b in av_twins[:i]])
```

Esto guarda una curiosa relación con la conjetura de Goldbach ya que si un número de esa lista puede escribirse como suma de dos números anteriores en la lista, entonces también puede escribirse de forma directa como suma de dos primos.. de dos maneras distintas!

$ 30 = 12 + 18 \\
30 = 11 + 19 \ (primos) \\
30 = 13 + 21 \ (primos)
$

Todo número que es media de dos primos gemelos (como el 30) es par, y por la conjetura de Goldbach ya "sabíamos" que puede ser escrito como suma de dos números primos. Pero... ¿son además esos dos números primos siempre parte de la lista de primos gemelos?
